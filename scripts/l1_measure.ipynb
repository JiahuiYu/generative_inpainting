```python
from pathlib import Path

import numpy as np
import matplotlib.pyplot as plt
import os.path as op
import cv2
import os
```

```python
for i, filename in enumerate(sorted(Path(data_pth).glob('*.npy'))):
    print(i, filename)
    
    spec = np.load(filename)
    print(spec.shape)
    h, w, _ = spec.shape
    spec = spec.reshape((h, w))
    print(spec.shape)
    plt.figure()
    plt.imshow(spec, 'gray')
    break
```
